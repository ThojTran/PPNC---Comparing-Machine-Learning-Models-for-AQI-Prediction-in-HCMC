```python
# MAPE: Mean absolute Percentage Error
METRICS = ['MAE', 'MSE', 'RMSE', 'R-Squared', 'MAPE'] 
```

```python
def evaluate_model_performance(y_true, predictions, label='dt'):
    mae = mean_absolute_error(y_true, predictions)
    mse = mean_squared_error(y_true, predictions)
    rmse = sqrt(mean_squared_error(y_true, predictions))
    r2 = r2_score(y_true, predictions)

    sum_y_true = np.sum(y_true)
    MAPE = np.mean(np.abs(y_true - predictions) / y_true)
     
    results = {label: [mae, mse, rmse, r2, MAPE]}
    
    return pd.DataFrame(results, index=METRICS).round(3)
```

```python
# from statsmodels.tsa.stattools import adfuller
def test_for_stationarity_single(series_data=None):
    results = adfuller(series_data.dropna())
    print(f'p-value: {results[1]}')    
    if results[1] < 0.05:  # if p value is less than 0.05, it is assumed to be stationary
        print('Series is stationary')
    else:
        print('Series is not stationary')


def plot_corr(corr, title='Feature Correlations with NO₂'):
    # Tạo mask mới chỉ che phần trùng lặp phía trên đường chéo
    mask = np.triu(np.ones_like(corr, dtype=bool), k=1)  # k=1 để giữ lại đường chéo chính
    
    # Thiết lập figure với kích thước phù hợp
    plt.figure(figsize=(12, 10))
    
    # Vẽ heatmap với các thông số được tối ưu
    sns.heatmap(
        corr, 
        mask=mask,
        cmap='Blues',
        annot=True,
        fmt='.2f',
        square=True,
        linewidths=.5,
        annot_kws={"size": 10}
    )
    
    # Điều chỉnh tiêu đề và nhãn
    plt.title(title, fontsize=16, pad=20)
    plt.xticks(fontsize=10, rotation=45, ha='right')
    plt.yticks(fontsize=10)
    
    # Đảm bảo layout phù hợp
    plt.tight_layout()
    
    
def visualize_series_single(series_data=None, days=None, title='NO₂ Pollutant AQI'):
    if series_data is not None:
        if days is None:
            series_data.plot(title=title)
        elif days > len(series_data) * 24:
            series_data.plot(title=title)
        else:
            series_data[-24 * days:].plot(title=title)


# from statsmodels.tsa.seasonal import seasonal_decompose
def decompose_series_single(series_data=None, model='additive', period=24):
    """
      model: {'additive' | 'multiplicative'}. 
          NOTE: 
            Use 'additive' if seasonal variation are relatively constant over time. i.e. linear.
            Use 'multiplicative' if the trend and seasonal variation increase or decrease over time.
    """

    if series_data is not None:
        results=seasonal_decompose(series_data, model=model, period=period)
        results.plot()
        return results


def series_to_supervised_dataset_single(series_data, n_lags=24):
    dataframe = pd.DataFrame()
    for i in range(n_lags, 0, -1):
        dataframe[f"{series_data.name.replace('.', '')}t-{i}"] = series_data.shift(i)
    final_data = pd.concat([dataframe, series_data], axis=1)
    final_data.dropna(inplace=True)
    return final_data


# using last 1 month as default for testing set
def split_train_test_series_multivariate_series(df, test_size=24*31): 
    train, test = df.iloc[:-test_size], df.iloc[-test_size:]
    X_train, y_train = train.iloc[:, :-1], train.iloc[:,-1]
    X_test, y_test = test.iloc[:, :-1], test.iloc[:,-1]

    return (X_train, y_train, X_test, y_test)
  

# using last 1 month as default for testing set
def split_train_test_series_single(series_data, test_size=24*31): 
    train, test = series_data[:-test_size], series_data[-test_size:]
    return (train, test)


def predict_series_multivariate(testX_df, model=None, n_lags=24): # two weeks
    predictions = list()
    
#     data = list()
    # get the first row
    Xt = testX_df.iloc[0].values.reshape(1,n_lags)
    print(Xt.shape)
    for i in range(0, len(testX_df)):
        yhat = model.predict(Xt)
        # yhat =  Xt[-1] + 20
        predictions.append(yhat[0])
        
        Xt = np.append(Xt[:,1:], yhat[0]).reshape(1,n_lags)
        # data.append(Xt)
    
    predictions = pd.Series(predictions, index=testX_df.index, name='NO₂')
    # return np.array(data), predictions
    return predictions
```

### 4. Thành phần nhiễu (Residual)
Thành phần residual có nhiều điểm nhiễu lớn vào cuối tháng 11 và giữa tháng 12, trùng với các đợt "spikes" trong biểu đồ gốc.

Một số đoạn residual về gần 0 hoặc thiếu dữ liệu (ví dụ giữa tháng 12), cho thấy gián đoạn đo đạc.

Nhiễu có thể đến từ:
- Phát thải bất thường (sự cố công nghiệp, xử lý chất thải đột ngột).
- Thời tiết đặc biệt (gió lặng, sương mù khiến khí không khuếch tán).
- Lỗi cảm biến hoặc gián đoạn truyền dữ liệu.

## Dự báo nồng độ NO₂ bằng mô hình XGBoost Regressor – TP.HCM

### Cấu hình dữ liệu:

### Diễn giải biểu đồ kết quả
Đường màu xanh lam (Train): phần dữ liệu được dùng để huấn luyện mô hình (25/12/2024 đến đầu 5/1/2025).

Đường màu đỏ (Test): dữ liệu thực tế dùng để kiểm tra mô hình (cuối ngày 4/1 đến 5/1/2025).

(Lưu ý: Biểu đồ không hiển thị rõ đường dự báo – prediction màu đen như mô tả mẫu, nên phần đánh giá dưới giả định đường dự báo đã được tính toán.)

## Đánh giá tổng quan
### 1. Khả năng bám sát xu hướng thực tế
Dữ liệu huấn luyện và kiểm tra thể hiện tính chu kỳ rõ ràng, với dao động mạnh trong ngày.

Mô hình có thể nắm bắt tốt cấu trúc chuỗi thời gian, như:
- Dao động hàng ngày (do giao thông, công nghiệp).
- Mức nền ổn định trong ngày.

### 2. Độ trễ trong dự báo các đỉnh cao (giả định có Prediction)
Trong các chuỗi thực tế như NO₂, các đỉnh cao thường không được dự báo chính xác, nếu không có thông tin bổ sung về thời tiết, hoạt động công nghiệp, giao thông.

Hiện tượng "làm mượt" (underestimation) các đỉnh là phổ biến với mô hình máy học không tích hợp đặc trưng ngoại sinh.

### 3. Tính ổn định và khả năng khái quát
Dữ liệu test (đỏ) biến động tương đối ổn định, không nhiễu bất thường → cho thấy mô hình được huấn luyện tốt.

Với chỉ 24 lags, mô hình đã học được chu kỳ ngắn hạn theo giờ → phù hợp cho dự báo 1 ngày tới.

### 4. Ưu điểm của XGBoost
Tận dụng tốt các mối quan hệ phi tuyến trong dữ liệu NO₂.

Vượt trội hơn các mô hình tuyến tính trong môi trường biến động ngắn hạn như đô thị.

Thời gian huấn luyện nhanh, dễ tinh chỉnh.



# XGBoost Regression

```python
# Tạo features và target
def create_features(data, lags):
    X, y = [], []
    for i in range(len(data) - lags):
        X.append(data[i:i+lags])
        y.append(data[i+lags])
    return np.array(X), np.array(y)

# Tạo dữ liệu huấn luyện với độ trễ
X_train, y_train = create_features(train_no2.values, lags=24)

# Chuyển đổi dữ liệu để phù hợp với XGBoost
dtrain = xgb.DMatrix(X_train, label=y_train)

# Tham số cho XGBoost
params = {
    'objective':'reg:squarederror',
    'n_estimators': 300,
    'learning_rate': 0.1,
    'max_depth': 8,
    'subsample': 0.6,
    'colsample_bytree' : 1.0,
    'random_state' : 0,
    'reg_lambda' : 0.5,
    'reg_alpha' : 0.5,
    'min_child_weight' : 3,
    'gamma' : 0.3,
    'eval_metric' :'rmse',
    'seed' : 42
}
    
```

```python
# Huấn luyện mô hình
num_rounds = 100
model = xgb.train(params, dtrain, num_rounds)
```

```python
# Dự báo cho ngày tiếp theo (sau bộ test)
last_sequence = pd.concat([train_no2, test_no2]).values[-24:]  # Lấy 24 giờ cuối cùng từ toàn bộ dữ liệu
predictions = []

for _ in range(steps):  # Dự đoán cho 24 giờ tiếp theo (1 ngày)
    # Reshape để phù hợp với yêu cầu đầu vào của mô hình
    current_input = xgb.DMatrix(last_sequence.reshape(1, -1))
    # Dự đoán giá trị tiếp theo
    next_pred = model.predict(current_input)[0]
    # Thêm dự đoán vào danh sách kết quả
    predictions.append(next_pred)
    # Cập nhật chuỗi (loại bỏ giá trị đầu và thêm dự đoán)
    last_sequence = np.roll(last_sequence, -1)
    last_sequence[-1] = next_pred

# Tạo index cho ngày tiếp theo sau bộ test
next_day_index = pd.date_range(start=test_no2.index[-1] + pd.Timedelta(hours=1), periods=steps, freq='H')
    
# Chuyển đổi dự báo thành Series với chỉ mục đúng
preds_xgb = pd.Series(predictions, index=next_day_index)
```

```python
# Vẽ biểu đồ
plt.figure(figsize=(12,2))
train_no2[-24*7:].plot(color='blue', label='train')
test_no2.plot(color='red', label='test')
preds_xgb.plot(color='black', label='predictions')
plt.legend(loc='upper left')
plt.title('NO2 1-Day using XGBoost Regression with 24-lags [TPHCM]\n')
```
